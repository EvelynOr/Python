#### Análisis realizado con Sweetviz

Es una libreria de Python que genera visualizaciones de alta densidad para realizar un EDA con sólo dos líneas de código. La salida es una aplicación HTML totalmente autónoma. El sistema se basa en la visualización rápida de valores objetivo y la comparación de conjuntos de datos. Su objetivo es ayudar al análisis rápido de las características del objetivo, datos de entrenamiento frente a datos de prueba y otras tareas de caracterización de datos.

#### Algunas de las características de SweetViz
+ Muestra cómo un valor objetivo (características dependientes) se relaciona con otras características.
+ Integra sin problemas asociaciones para tipos de datos numéricos (correlación de Pearson), categóricos (coeficiente de incertidumbre) y categórico-numéricos (relación de correlación), para proporcionar la máxima información para todos los tipos de datos.
+ Detecta automáticamente características numéricas, categóricas y de texto, con anulaciones manuales opcionales.
+ Información resumida sobre el tipo, valores únicos, valores que faltan, filas duplicadas, valores más frecuentes.
+ Análisis numérico como mín/máx/rango, cuartiles, media, moda, desviación estándar, suma, desviación absoluta mediana, coeficiente de variación, curtosis, asimetría (analyticsvidhya).
