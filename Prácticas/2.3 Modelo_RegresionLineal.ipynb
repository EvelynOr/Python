# Modelo de Regresión Lineal

Las relaciones lineales, lo que quiere decir que cuando se grafican, los puntos se aproximan al patrón de una línea recta.

Si se concluye que existe una correlación lineal entre 'x' y 'y', se puede obtener una ecuación lineal que exprese 'y' en términos de 'x', y la ecuación puede emplearse para predecir valores de 'y' a partir de valores dados de 'x'. Triola, M. F. (2009).

El resultado de la regresión lineal es una función lineal que predice la variable de respuesta (dependiente 'y') en función de la variable predictora (independiente 'x').

#### Objetivo: 

predecir el precio de un carro utilizando variables o datos que se tienen. Es una estimacion de cuánto podria ser el precio.

## 1.Desarrollo de modelos

Un modelo o estimador es una ecuación matemática utilizada para predecir un valor con base a datos dados [1]. Relaciona una o más variables o características independientes con variables dependientes. Por lo general, cuantos más datos relevantes tenga, más preciso será el modelo.

En este caso, si ingresa las millas por galón en carretera de un modelo de automóvil como variable o característica independiente, la salida del modelo o variable dependiente es el precio. Si se ingresa múltiples variables o características independientes en el modelo puede predecir un precio más preciso para el automóvil. 

Se revisa la regresión lineal simple, la regresión lineal múltiple y la regresión polinomial.

### 1.1 Regresión lineal simple 

Se refiere a una variable independiente para hacer predicciónes. Es un método que  ayuda a entender la relación entre dos variables continuas. 

La variable independiente, predictor 'x' y la variable dependiente del objetivo 'y'.

El parámetro b0 es el intercepto = valor de Y con X = 0

El parámetro b1 es la pendiente o 'm' o coeficiente de regresion = el valor en que cambia Y cuando X incrementa en 1




Cuando se ajusta o entrena el modelo, se obtienen los parámetros los cuales se calculan usando el método de mínimos cuadrados y se obtiene

Funcion lineal:

Yhat = a + bX                                       [2]

En la regresión lineal simple, la gráfica de la ecuación de regresión se llama línea de regresión estimada


En este ejemplo se quiere ver cómo 'Highwaympg' puede ayudar a predecir el precio del automóvil. 

```python
# importar librerias 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# leer data y asignarla a un df [3]
df = pd.read_csv('automobileEDA.csv')
```

```python
# ver la data
df.head(6)
```

```python
# importar libreria para el modelo de regresion lineal
from sklearn.linear_model import LinearRegression
```

```python
# definir las variables para la prediccion [4]
X = df[['highwaympg']]
Y = df['price']
```

```python
# crear el objeto para la regresion lineal = asignar el modelo a una variable
lm1=LinearRegression() 
lm1
```

```python
# ajustar el modelo de regresion = obtener parámetros
lm1.fit (X, Y) 
```

```python
# intercepto (bo o a)
print(lm1.intercept_)

# pendiente (b1 o m)
print(lm1.coef_)
```

modelo de regresion: Y = bo + b1X              -----------------o----------------                              Y = a + mX

Price = 38,423.31 - 821.73highwaympg


INTERPRETACION: Price =  38,423.31 - 821.73 veces millas de carretera por galón


```python
# visualizar la data de las variables seleccionadas 
X,Y
```

```python
# prediccion 1, cuando  X = highwaympg = 27  [5]
Yhat1 =  lm1.intercept_ - 821.73*27
Yhat1
```

```python
# prediccion 2, cuando  X = highwaympg = 26
Yhat1 =  lm1.intercept_ - 821.73*26
Yhat1
```

```python
# prediccion 3, para todos los valores de X = highwaympg
Yhat1 = lm1.intercept_ - lm1.coef_ * X
Yhat1
```

```python
# prediccion 4, con todas las X en un rango de Y = [0:5]
Yhat1 = lm1.predict(X)
Yhat1[0:5]  
```

### 2. Evaluando el modelo con visualizacion

En la regresión lineal, la gráfica de la ecuación de regresión se llama línea de regresión estimada. 

ŷ es el valor estimado de y para un valor específico de x.

La gráfica de regresión son una buena estimación de:

-la relación entre dos variables

-la fuerza de la correlación 

-la dirección de la relación (positiva o negativa)


### 2.1 grafica-diagrama de dispersion

muestra los valores y posible relacion de dos variables de un conjunto de datos

permite estudiar cualitativamente, la relación entre ambas variables

```python
# x = nombre de la columna de la VAR independiente
# y = nombre de la columna de la VAR dependiente o destino 
# data, el nombre del marco de datos

# librerias
import seaborn as sns
import matplotlib.pyplot as plt

# tamno grafica
width = 12
height = 10
plt.figure(figsize=(width, height))

# grafica
sns.regplot (x = "highwaympg", y = "price", data = df)
plt.ylim (0,)                                              # Configura límite vertical
```

ANALISIS: la grafica muestra que el precio está negativamente correlacionado con "highwaympg"  ya que la pendiente de regresión es negativa.

Una cosa a tener en cuenta al mirar un gráfico de regresión es prestar atención a cuán dispersos están los puntos de datos alrededor de la línea de regresión. Esto le dará una buena indicación de la variación de los datos y si un modelo lineal sería el mejor ajuste o no. Si los datos están demasiado alejados de la línea, este modelo lineal podría no ser el mejor modelo para estos datos.

Comparemos este gráfico con el gráfico de regresión de price y "peakrpm".

```python
#gráfico de regresión de price y "peakrpm"

plt.figure(figsize=(width, height))
sns.regplot(x="peakrpm", y="price", data=df)
plt.ylim(0,)
```

Al comparar el gráfico de regresión de "peakrpm" y "highwaympg", se observa  que los puntos para "highwaympg" están mucho más cerca de la línea generada y con una tendencia negativa. 

Los puntos de "peakrpm" están más dispersos alrededor de la línea predicha y es mucho más difícil determinar si los puntos disminuyen o aumentan a medida que aumenta "peakrpm". Esto se puede verificar una matriz de correlacion entre ambas variables.


```python
# matriz de correlacion
df[["peakrpm","highwaympg","price"]].corr()
```

### 2.2 grafica de residuos, representa el error =  valor real - valor esperado

para examinar la bondad de ajuste en regresión y ANOVA

residuo (e) =  valor observado (y) -  valor pronosticado (Yhat) se llama residuo (e). 

Cuando observamos un gráfico de regresión, el residuo es la distancia desde el punto de datos hasta la línea de regresión ajustada.

Una gráfica de residuos muestra los residuos en el eje 'y vertical' y la variable independiente en el 'eje x horizontal'. Se deben observar la dispersión de los residuos:
- Si los puntos de residuos se distribuyen aleatoriamente alrededor del eje x, entonces un modelo lineal es apropiado para los datos.

¿Porqué es eso? La distribución aleatoria de residuos significa que la varianza es constante y, por lo tanto, el modelo lineal se ajusta bien a estos datos.

```python
# grafica  sencilla
sns.residplot (df['highwaympg'], df[ 'price'])
```

```python
# grafica mas elaborada
# 'highwaympg', VAR independiente x 
# 'price', VAR dependiente y

# libreria
import seaborn as sns

# tamno grafica
width = 12
height = 10
plt.figure(figsize=(width, height))

# grafica
sns.residplot (df['highwaympg'], df['price'])
plt.show()
```

ANALISIS: los residuos no se distribuyen aleatoriamente alrededor del eje x, lo que podria indicar que tal vez un modelo no lineal sea más apropiado para estos datos.

### 3. Regresión polinómica y Pipelines (canales de datos)

Se obtiene al elevar al cuadrado o establecer términos de orden superior en las variables predictoras del modelo

Es un caso especial de la regresión lineal general, cuando un modelo lineal no es el mejor ajuste para los datos, sirve para modelar relaciones no lineales. 

La regresión polinómica, se transforman los datos en un polinomio, luego se utiliza la regresión lineal para ajustarse al parámetro. 

Piplines = una forma de simplificar el código

```python
# ejemmplo: generar un modelo polinomico de tercer grado

#libreria requerida
import numpy as np

# definir variables
x = df['highwaympg']
y = df['price']

#generar modelo
f = np.polyfit(x, y, 3)
p = np.poly1d(f)                      # método de atributo polinomial. Ajuste por mínimos cuadrados

#ver modelo
print(p)
```

### 3.1 Uso de  piplines

The pipeline can involve pre-processing, feature selection, classification/regression, and post-processing.

La clase Pipeline de Scikit-learn está diseñada como una serie de transformaciones de datos, formada por un conjunto de modelos, que aplica estimadores con un resultado-estimador final. En resumen, organiza el flujo de datos y la salida.

La figura muestra el proceso de una prediccion con pipelines. [7]

```python
# librerias para la prediccion
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
# convertir los datos a tipo float [6]
Zp = df[['horsepower', 'curbweight', 'enginesize', 'highwaympg']]
Zp = Zp.astype(float)
```

```python
# crear una lista de tuplas [8]
Input=[('scale', StandardScaler()), ('polynomial', PolynomialFeatures (degree=2)), ('model', LinearRegression())]
pipe = Pipeline(Input)
```

flecha blue = nombre de los estimadores 

flecha roja = modulos para construir el modelo 



luego se asigna la lista a un Pipeline que se convierte en un pipeline object

```python
# entrenar el pipeline al asignar el metodo 'fit'al objeto creado 
pipe.fit(Zp, y)
             
```

```python
# hacer la prediccion
Yhat3=pipe.predict(Zp)
Yhat3[0:4]
```
